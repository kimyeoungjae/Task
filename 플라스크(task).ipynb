```python
from flask import Flask, Response, request, send_file, redirect#간단히 플라스크 서버를 만든다
import requests
import urllib.request
import json
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from io import BytesIO, StringIO

app = Flask(__name__)

@app.route("/tospring")
def spring():
    
    if request.method == 'GET':
        dic = request.args.get("js","모름")
        arry = json.loads(dic)
        print(type(arry))
        
        
        cameralist = []
        objectlist = []
        activitylist=[]
        feedlist=[]
        drinkinglist=[]
        for i in arry:
            cameralist.append(int(i["camera_id"]))
            objectlist.append(int(i["object_id"]))            
            activitylist.append(int(i["activity"]))
            feedlist.append(int(i["feed"]))
            drinkinglist.append(int(i["drinking"]))
        
        print(cameralist[0])
        print(objectlist[0])
        plt.figure(figsize=(20,4))
        plt.plot(activitylist, 'y', label = 'activity') ## 노란색       
        plt.plot(feedlist, 'r',label = 'feed' ) ## 빨간색
        plt.plot(drinkinglist, 'm', label = 'drinking')  ## 보라색   
        plt.legend()
        ax = plt.gca()
        ax.set_facecolor('black')
        plt.show()        
        plt.savefig('C:/Users/SMT056/Desktop/homework/Task/WebContent/img/save.{}.{}.png'.format(cameralist[0], objectlist[0]))
        value = ("{}.{}".format(cameralist[0],objectlist[0]))
        print(value)
        return redirect("http://localhost:8090/Task/Test.jsp?rs="+value)
    
if __name__ == '__main__':
    app.run(debug=False,host="127.0.0.1",port=5000)
```
